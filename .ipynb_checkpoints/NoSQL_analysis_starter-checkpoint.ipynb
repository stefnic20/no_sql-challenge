```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
limit = 20

#variable to hold list
results = list(establishments.find(query).limit(limit))

#print
pprint(results)
```

```python
# Use count_documents to display the number of documents in the result
establishments.count_documents({'scores.Hygiene' : 20})
```

```python
# Display the first document in the results using pprint
query = {'scores.Hygiene' : 20}
sort = {'scores.Hygiene', 1}
limit = 1

#variable to hold list
results1 = list(establishments.find(query).limit(limit))

#print
pprint(results1)
```

```python
# Convert the result to a Pandas DataFrame
Hygiene_20_df = pd.DataFrame(results)
# Display the first 10 rows of the DataFrame
Hygiene_20_df.head(11)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex':'London'},'RatingValue': {'$gt': 3}}          
results2 = list(establishments.find(query))

# Display the first document in the results using pprint
pprint(results2)
```

```python
# Use count_documents to display the number of documents in the result
count = establishments.count_documents({'LocalAuthorityName': {'$regex':'London'},'RatingValue': {'$gt': 3}})  
count
```

```python
# Convert the result to a Pandas DataFrame
LA_Ratings_df = pd.DataFrame(results2)

# Display the first 10 rows of the DataFrame
LA_Ratings_df.head(11)
```

```python
# Display the number of rows in the DataFrame
LA_Ratings_df.shape[0]
```
